# Compute central difference for a vector of data

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 30
x = np.linspace(0.1,3,n)
f = np.sin(x)
```

```python
plt.plot(x,f,'k')
plt.plot(x,f,'rx',linewidth=2)
```

```python
dx = x[1]-x[0]
dfdx = np.zeros(n)

dfdx[0] = (f[1] - f[0]) / (x[1] - x[0])
for i in range(1,n-1):
    dfdx[i] = (f[i+1] - f[i-1])/(x[i+1]-x[i-1])
dfdx[-1] = (f[-1] - f[-2]) / (x[-1]-x[-2])
```

```python
plt.figure()
plt.plot(x, np.cos(x),'o', label='true derivative')
plt.plot(x,dfdx, 'rx', label = 'computed derivative')
plt.legend()
```
