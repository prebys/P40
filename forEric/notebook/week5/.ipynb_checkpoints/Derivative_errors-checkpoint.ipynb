# let's compare the errors for central difference and forword difference

```python
from math import exp, sin, cos
```

```python
# this is the function f(x) = exp(sin(2x))

def f(x):
    return exp(sin(2*x))

# this is the exact derivative of the function f(x) = exp(sin(2x))
def fprime(x):
    return 2*exp(sin(2*x))*cos(2*x)
```

```python
# forword difference
def calc_fd(f,x,h):
    fd = (f(x+h) - f(x))/h 
    return fd
# central difference
def calc_cd(f,x,h):
    cd = (f(x+h/2) - f(x-h/2))/h 
    return cd
```

```python
x = 0.5
an = fprime(x)

# defining an array of different step "h"
hs = [10**(-i) for i in range(1,12)]

# computing the difference between the exact derivative and the forword difference
fds = [abs(calc_fd(f,x,h) - an) for h in hs] 

# computing the difference between the exact derivative and the central difference
cds = [abs(calc_cd(f,x,h) - an) for h in hs]
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
print("h abs. error in fd abs. error in cd") 
for h,fd,cd in zip(hs,fds,cds):
    print(rowf.format(h,fd,cd))
h
```

```python
from matplotlib import pylab as plt
plt.plot(hs,cds,'r-')
plt.plot(hs,fds,'b-')
plt.plot(hs,cds,'ro',label = 'central diff')
plt.plot(hs,fds,'bo',label = 'forword diff')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('h')
plt.ylabel('abs error')
plt.legend()
```
