```python
def powers(a, n):
    xx = [a**i for i in range(0,n+1)]
    return xx

xx = powers(2,4)
print(xx)
```

```python
def recur(n,x):
    a = np.array([]) 
    for i in range(0,n):
        a= np.append(a, x)
        x = 2 * x +1
    return a
        
        
```

```python
ss = recur(5,1)
print(ss)
```

```python
UPPER = 2
LOWER = 0
tau = 2*np.pi
x = np.linspace(LOWER,UPPER,100)
s = np.sin(tau*x)
c = np.cos(tau*x) 
plt.plot(x,s,"b-",label="sin") 
plt.plot(x,c,"r-",label="cos") 
plt.xlabel("x") 
plt.ylabel("y")
plt.title("Two Periods of a Sine and Cosine") 
plt.legend(frameon=False)
plt.show()
```

```python
UPPER = -5
LOWER = 5
tau = 2*np.pi
x = np.linspace(LOWER,UPPER,100)
s = np.sinc(tau*x)
plt.plot(x,s,"b-",label="sinc") 
plt.xlabel("x") 
plt.ylabel("y")
plt.title("Sinc") 
plt.legend(frameon=False)
plt.show()
```

```python
def logmap(p, r):
    p1 = r * p *(1-p)
    return p1

r = 3.0 
p0 = 0.1 
p1 = logmap(p0,r)
p2 = logmap(p1,r)
p3 = logmap(p2,r)
print(p1,p2,p3)
```

```python
def buildarrsy(n,p,r):
    a = np.array([]) 
    for i in range(0,n):
        a= np.append(a, p)
        p = logmap(p, r)
    return a
a = buildarrsy(6,0.1,3.0)
print(a)
```

```python
y = buildarrsy(40,0.1,2.8)
x = np.arange(40)
plt.title('Logistic Map 2.8')
plt.plot(x,y,'-r')
```
