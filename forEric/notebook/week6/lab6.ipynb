Numerical Differentiation

```python
def right(f,x,h):
    return (f(x+h)-f(x))/h
def center(f,x,h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
def f(x): # derivative 0 
    return 2
def g(x): # derivative 3 
    return 3*x
def h(x): # derivative 4x 
    return 2*x**2
   
```

```python
print(right(f,1,0.01)) 
print(right(g,1,0.01)) 
print(right(h,1,0.01))
```

```python
 
print(right(f,4,.1))
print(right(g,4,.1))
print(right(h,4,.1))
```

```python
def f(x): # derivative 6x**2
    return 2*x**3
print("right:", right(f,1,0.1), "center:", center(f,1,0.1)) 
print("right:", right(f,1,0.01), "center:", center(f,1,0.01))
print("right:", right(f,1,0.001), "center:", center(f,1,0.001))
```

```python
def f(x):
    return 0.5*x**2
x = np.linspace(0,1,100) 
y = center(f, x, 0.1) 
plt.plot(x,y,"-b") 
plt.xlabel("x") 
plt.ylabel("y")
plt.show()
```

```python
def f(x):
    return x**3
x = np.linspace(-2,2,100) 
y = center(f, x, 0.1) 
plt.plot(x,y,"-b",label="f'(x)") 
plt.plot(x,f(x),"-r",label="f(x)") 
plt.xlabel("x") 
plt.ylabel("y")
plt.legend(ncol=1)
plt.show()
```

```python
def f(x):
    return np.sin(x)

x = np.linspace(0,2*np.pi,100) 
y = center(f, x, 0.01) 
plt.plot(x,y,"-b",label="f'(x)") 
plt.plot(x,f(x),"-r",label="f(x)") 
plt.plot(x,np.cos(x),"xg",label="cos(x)") 
plt.xlabel("x") 
plt.ylabel("y")
plt.legend(ncol=1)
plt.show()
```

Euler method

```python
def euler(dt, x, y, vx, vy, ax, ay): 
    x = x + vx * dt
    y = y + vy * dt
    vx = vx + dt * ax
    vy = vy + dt * ay
    return x, y, vx, vy
```

```python
print(np.around(euler(0.134, 0.659, 0.282, 0.662, 0.643, 0.900, 0.451),2))
print(np.around(euler(0.924, 0.959, 0.575, 0.299, 0.710, 0.699, 0.471),2))
euler(0.134, 0.659, 0.282, 0.662, 0.643, 0.900, 0.451)
```

```python
tau = 2*np.pi
vi =20 #[m/s]
g = 9.8 # [m/s^2]
theta = tau/8
dt = 0.01 # [s] 
x=0 #[m] 
y=0 #[m]
vx = vi*np.cos(theta) 
vy = vi*np.sin(theta)
ax = 0
ay = - g
t = 0
```

```python
tjx = []
tyx = []
vjx = []
vyx = []
tx = []
while y>=0:
    t = t + dt 
    tx.append(t)
    tjx.append(x)
    tyx.append(y)
    vjx.append(vx)
    vyx.append(vy)
    x,y,vx,vy = euler(dt,x,y,vx,vy,ax,ay)
```
