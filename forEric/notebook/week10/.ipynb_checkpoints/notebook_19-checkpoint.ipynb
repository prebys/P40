```python
# instead of creating manually all these info let's save all these info in the class directly 
# let's create a special initialization 
# we are creating a methos inside a class

class Employee:
    # this is an init method
    def __init__(self, first, last, pay):
        self.first = first  # instance variable
        self.last = last    # instance variable
        self.pay = pay      # instance variable
        self.email = first+ '.' + last + '@company'

    # this is a full name method
    def fullname(self):
        return 'emplyee name: {} {}'.format(self.first, self.last)
        
```

```python
emp_1= Employee('bob','red',50000)
emp_2= Employee('carl','red',60000)
```

```python
print(emp_1.first)
print(emp_2.email)
```

```python
# let's use the full name method we created 
print(emp_1.fullname())
```
