```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
# exercise 10.1
T0 = 293  # K
M0 = 4.65e-26 # kg
kb0 = 1.38e-23 # J/K

M = 1 # mass of gas particle
T = 1 # temperature of gas 
kb = 1 # Bolzman constant

def mbspeed(v):
    return (M*v / (kb*T)) * np.exp(-M*v**2/(2*kb*T))
```

```python
vvec = np.arange(0,5,.1)
T = 1
plt.plot(vvec,mbspeed(vvec),'-r',label='T = 1')
T = 3
plt.plot(vvec,mbspeed(vvec),'-g',label='T = 3')
T = 5
plt.plot(vvec,mbspeed(vvec),'-b',label='T = 5')
plt.legend()
plt.xlabel('speed (V0)')
```
