# Using matrix to solve linear equation 
(you will see mor eof this in  Phy80 class)

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
# Choose the "true" parameters.
m_true = -0.9594
b_true = 4.294
f_true = 0.534
```

```python
# Generate some synthetic data from the model.
N = 50
x = np.sort(10 * np.random.rand(N))
yerr = 0.1 + 0.5 * np.random.rand(N)
y = m_true * x + b_true
y += np.abs(f_true * y) * np.random.randn(N)
y += yerr * np.random.randn(N)
plt.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)

# let's plot the correct solution
x0 = np.linspace(0, 10, 500)
plt.plot(x0, m_true * x0 + b_true, "k", alpha=0.3, lw=3)
plt.xlim(0, 10)
plt.xlabel("x")
plt.ylabel("y");
```

```python
# Create the design matrix
X = np.vstack([x, np.ones(len(x))]).T

# Solve for coefficients using least squares
coefficients, residuals, rank, singular_values = np.linalg.lstsq(X, y, rcond=None)

# Extract slope and intercept
slope, intercept = coefficients

print("Slope:", slope)
print("Intercept:", intercept)

plt.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
plt.plot(x0, m_true * x0 + b_true, "k", alpha=0.3, lw=1)
plt.plot(x0, slope * x0 + intercept, "b", alpha=.3, lw=1)
plt.xlim(0, 10)
plt.xlabel("x")
plt.ylabel("y");
```

# Weather system 
(using matrix to predicts a weather system)

let's show how we can use matrix and vectors to study a weather system

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# this is my probability transition matrix for our weather model
# The states are Raining, Nice or Cloudy 
# the probability for tomorrow weather is that if today is raining, 
# tomorrow there is 50% probability to rain, 25% probability the weather will be nice or cloudy 
# and so on ...
A = np.array([[0.5,0.5,0.25],[0.25,0.0,0.25],[0.25,0.5,0.5]])
```

```python
print('Raining  Nice  Cloudy')
print(A)
```

```python
#today is Raining
xtoday = np.array([[1],[0],[0]])
print(xtoday)
```

```python
# tomorrow will be ... 
print(A @ xtoday)
```

```python
# Let's compute the probability in 50 days form now
for k in range(50):
    xtomorrow = A @ xtoday
    print(k)
    print(xtomorrow)
    xtoday = xtomorrow
```

it will converge which ever is the initial status

```python
# let's save this states in time
xtoday = np.array([[1],[0],[0]])
A = np.array([[0.5,0.5,0.25],[0.25,0.0,0.25],[0.25,0.5,0.5]])
print(xtoday)
the_weather = np.zeros((3,50))
the_weather[:,0] = xtoday[:,0]
for k in range(50):
    xtomorrow = A @ xtoday
    the_weather[:,k] = xtomorrow[:,0]
    xtoday = xtomorrow
    print(the_weather[2])
```

```python
# let's plot this probability as function of time

plt.plot(the_weather.transpose()[:,0], 'r:', linewidth=3, label='Rainy')
plt.plot(the_weather.transpose()[:,1], 'k--', label='Nice')
plt.plot(the_weather.transpose()[:,2], 'b-', label='Cloudy')
plt.grid(True)
plt.xlabel('days')
plt.ylabel('Probabilty')
plt.legend()
```
