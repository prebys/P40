# Error for central and forword derivative as a function of step

```python
from math import exp, sin, cos
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# let's define a function and his derivative
def f(x):
    return exp(sin(2*x))
def fprime(x):
    return 2*exp(sin(2*x))*cos(2*x)
```

```python
# let's define the central derivative and the forword derivative    
def calc_fd(f,x,h):
    fd = (f(x+h) - f(x))/h 
    return fd

def calc_cd(f,x,h):
    cd = (f(x+h/2) - f(x-h/2))/h 
    return cd
```

```python
# let's fix a point and compute derivative , 
#forword finite derivative and central finite derivative of the function
x = 0.7
an = fprime(x)
#Let's  define a range of steps 
hs = [10**(-i) for i in range(1,12)] 

fds = [abs(calc_fd(f,x,h) - an) for h in hs] 
cds = [abs(calc_cd(f,x,h) - an) for h in hs]
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
print("h abs. error in fd abs. error in cd") 
for h,fd,cd in zip(hs,fds,cds):
        print(rowf.format(h,fd,cd))
```

```python
# let's plot the error as a function of the steps

plt.plot(hs,fds,'-r',label='forward')
plt.plot(hs,fds,'or')
plt.plot(hs,cds,'-b', label='central')
plt.plot(hs,cds,'ob')
plt.yscale('log')
plt.xscale('log')
plt.ylabel('error')
plt.xlabel('h')
plt.legend()
```

- the minimum error is smaller for central derivative
- the minimum eroor for the forword derivative is reached for a smaller step
