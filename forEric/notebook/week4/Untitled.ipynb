```python
import numpy as np 
def find_max(seq):
    max_item = 0
    for item in seq:
        if item > max_item:
            max_item = item
    return max_item

seq = [1,3,4,5,6,5,4,5,6,7,6,5,4,3,7,6,5,6,7,8,9,8,7,44,5,6,7,8,9,9,9,9,9,9,8,7,7,6,5,5,4,5,3,6,7,9,22]
find_max(seq)
```

```python
import time
st = time.time()
_max = find_max(seq)
et =time.time()
print(_max)
print("%s seconds " % (et-st))
```

```python
def palindrome(string):
    if len(string)< 2:
        return True
    elif string[0]!=string[-1]:
        return False
    return palindrome(string[1:-1])

palindrome('qaciicaq')
```
