# Histogram
Histograms divide the data up into bins and plots the number of points in each bin. First create some random data using np.random.normal.

```python
#Example: 
x = np.random.normal(10,size=200)
print(x)
```

```python
h, bins = np.histogram(x, bins=10)
print(h)
plt.plot(np.arange(len(h)),h,'-',drawstyle ='steps')
```

```python
plt.hist(x, bins='auto')
```

```python
# let's make an image 100 x 100 with number between 0 and 1 
a = np.random.random((100,100)) 
#let's give a higher value to some of the ccd pixels
a[2:5,3:6] = 10
print(a)
```

```python
# we can use imshow command to plot the image
import matplotlib.pyplot as plt 
plt.imshow(a, vmin=0,vmax=6)
```

```python
#We can "flatten" the two dimensions array in one dimension array to make an histrogram of pixels values distributions 
b= a.flatten()
print(b)
plt.hist(b)
plt.yscale('log')
```

let's make a "mask" to the image finding all these pixels that have value larger than 2
in astronomy sometimes we use 'masks' to find outliers pixels (cosmis rays, bad pixels of the CCD, ecc)

mask is now an 2D arrays of the same shape of our image but a bolean array with False for good pixels and True for "bad" pixels 

```python
mask = a>2
print(mask)
```

```python
# let's now replace the bad pixels of the image with the median of the image
a[mask]= np.median(a)
plt.imshow(a,vmin=0,vmax=6)
```

# numpy vs for loop
Let's make an example to see that numy is much faster then loop to do operations 

```python
# lets define a large array
x = np.arange(10000000)

# get the start time
st = time.time()

# main program
# find sum to first 10 million numbers
sum_x = 0
for i in x:
    sum_x += i

# wait for 3 seconds
#time.sleep(3)
print('Sum of first 1 million numbers is:', sum_x)

# get the end time
et = time.time()

# get execution time
res = et - st
print('CPU Execution time:', res, 'seconds')
```

Let's do the same in numpy

```python
# get the start time
st = time.time()

# main program
# find sum to first 10 million numbers
sum_x = np.sum(x)

# wait for 3 seconds
#time.sleep(3)
print('Sum of first 1 million numbers is:', sum_x)

# get the end time
et = time.time()

# get execution time
res = et - st
print('CPU Execution time:', res, 'seconds')
```

```python
from matplotlib import pylab as plt
# let's define x axis 
x = np.linspace(0, 1, 20)
# let's add some random scatter
rng = np.random.default_rng()
# we add the random scatter to the cos function 
y = np.cos(x) + 0.3*rng.random(20)
```

```python
plt.plot(x, y, 'o')
```

```python
# let's use the poly1d fit function in numpy to fit the points
p = np.poly1d(np.polyfit(x, y, 3))
```

```python
# what is p ?
# let's try 
print(p(0.4))
print(p(0.3))
```
