# Compare forward, backward and central differences for numerically computing derivatives

```python
import numpy as np
from matplotlib import pyplot as plt
```

let's define a function f= sin(t) in the range -2,40  

```python
dt = .4
t = np.linspace(-2,4,60)
f = np.sin(t)
```

```python
# exact derivative
dfdt = np.cos(t)
```

```python
# plotting command
plt.plot(t,f,'k--', linewidth=1.2)
plt.grid(True)
plt.plot(t,dfdt,'k',linewidth=3)
plt.legend(['Function','Exact Derivative'],fontsize=14)
plt.xlim(-2,4)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
# Forward Difference Approximation
dfdtF = (np.sin(t + dt) - np.sin(t)) / dt
```

```python
# Backward Difference Approximation
dfdtB = (np.sin(t) - np.sin(t - dt)) / dt
```

```python
# Central Difference Approximation
dfdtC = (np.sin(t + dt) - np.sin(t - dt)) / (2*dt)
```

```python
plt.figure(figsize=(20,4))
plt.subplot(1,2,1)
plt.plot(t,dfdt,'k',linewidth=3)
plt.plot(t,dfdtF,'b',linewidth=1.2) # Forward
plt.plot(t,dfdtB,'g',linewidth=1.2)# Backword
plt.plot(t,dfdtC,'r',linewidth=1.2) # Central
plt.legend(['Exact Derivative','Forward','Backward','Central'],fontsize=14, framealpha=1)
plt.grid(True)

# 
plt.subplot(1,2,2)
plt.plot(t,dfdt,'k',linewidth=3)
plt.plot(t,dfdtF,'b',linewidth=1.2) # Forward
plt.plot(t,dfdtB,'g',linewidth=1.2)# Backword
plt.plot(t,dfdtC,'r',linewidth=1.2) # Central
plt.grid(True)
plt.xlim(0.5,1.5)
plt.ylim(0,1)
plt.show()
```
