# numerical integration with the rectangle and trapezoidal rules
The True value is  
$$\int_{0}^{10} sin(x)\,dx$$ = 1.8391
and we will see how we can approximate this integral below.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = 0
b = 10
dxf = 0.01
xf = np.linspace(a,b, int(b/dxf))
yf = np.sin(xf)
plt.figure()
plt.grid(True)
plt.plot(xf,yf)
```

```python
dxc = 0.5
xc = np.linspace(a,b, int(b/dxc))
edges = np.linspace(a,b,int(b/dxc)+1,endpoint=True)
yc = np.sin(xc)
plt.plot(xf,yf)
plt.stairs(yc,edges,color='r')
plt.grid(True)
plt.show()
```

```python
n = len(xc)
# left-rectangle rule
area1 = 0 
for i in range(n-1):
    area1 = area1 + yc[i] * dxc
print(area1)
```

```python
# right-rectangle rule
area2 = 0 
for i in range(n-1):
    area2 = area2 + yc[i+1] * dxc
print(area2)
```

```python
# trapezoid rule 
area3 = 0 
for i in range(n-1):
    area3 = area3 + (yc[i] + yc[i+1]) * (dxc/2)
print(area3)
```

```python
# we can use build in python function 
area1 = sum(yc[:-1]) * dxc
area2 = sum(yc[1:]) * dxc
area3 = np.trapz(yc,xc)  # this will get very close to the real case
print(area1, area2, area3) 
area3 = np.trapz(yc) * dxc
print(area1, area2, area3) 
```

```python
# we can use build in python function 
area1f = sum(yf[:-1]) * dxf
area2f = sum(yf[1:]) * dxf
area3f = np.trapz(yf,xf)  # this will get very close to the real case
print(area1f, area2f, area3f) 
area3 = np.trapz(yf) * dxf
print(area1f, area2f, area3f) 
```
