```python
def f(x):
    return 5*(x-3)

def fp(x):
    return 5
```

```python
for x in [-1,3,4]: 
    print(f(x), fp(x))
```

```python
def bisection(f, a, b):
    c = (a + b)/2
    if f(a)*f(c)<= 0:
        an = a
        bn = c
    else: 
        an = c
        bn = b
    return an, bn # updated values
```

```python
print(bisection(f,0,4)) 
print(bisection(f,3,4))
```

```python
def newton(f,fp,x): 
    xn = x - f(x)/fp(x)
    return xn # updated value
```

```python
for x in [-100,5,1000]: 
    print(newton(f,fp,x))
```

```python
def secant(f,xnm1,xn):
    xnp1 = xn - f(xn) * (xn - xnm1) / (f(xn)-f(xnm1))
    return xn,xnp1
```

```python
print(secant(f,0,4)) 
print(secant(f,5,3))
```
