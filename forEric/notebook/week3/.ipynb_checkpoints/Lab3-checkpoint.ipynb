```python
# 3.7 
def fibonaci(n1,n2,nterm):
    fiblist=[]
    count = 0
    if n1==0:
        nth = 0
    elif n2==1:
        nth= 1
    while count < nterm:
            fiblist.append(n1)
            nth = n1+n2
            n1=n2
            n2=nth
            count = count + 1
    return fiblist
```

```python
def arith(n):
    sum = 0
    a = 1
    d = 1
#    Sn= (n+1) * (a+(a+n*d)/2)
    for j in range(0,n):
        sum = sum + ((j+1) * ((a+(a+j*d))/2))
    return sum

print(arith(0))
print(arith(1))
print(arith(2))
```
