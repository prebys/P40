# Accessing system information



```python
import platform
import os
import sys

def system_info():
    print("=== Platform Information ===")
    print("System:", platform.system())
    print("Release:", platform.release())
    print("Version:", platform.version())
    print("Machine:", platform.machine())
    print("Processor:", platform.processor())

    print("\n=== Python Information ===")
    print("Python Version:", sys.version)
    print("Executable:", sys.executable)

    print("\n=== Environment Variables ===")
    for key, val in os.environ.items():
        print(f"{key}: {val}")

system_info()
```

# Getting directory and file information

```python
import os

for name in os.listdir("."):  # list files in the directory
    print(name)
```

```python
# recursively go to subdirectories

for root, dirs, files in os.walk("."):  # This will recursively go throuh all subdirectories
    print("Directory:", root)
    for f in files:
        print("  File:", f)
```
