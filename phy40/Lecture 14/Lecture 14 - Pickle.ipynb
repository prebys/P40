# Pickle Demonstration

Pickle is used to store and retrieve data of any type in Python


```python
import pickle

# Example object â€” can be almost anything: dicts, lists, custom classes, etc.
data = {
    "name": "Alice",
    "age": 30,
    "scores": [95, 88, 76]
}

# ---- Write to file ----
f=open("data.pkl", "wb")   # 
pickle.dump(data, f)
```

```python
# ---- Read back from file ----
f = open("data.pkl","rb")
loaded_data = pickle.load(f)

print("Loaded:", loaded_data)
```

# Multiple Objects

There are many ways to do this, but the simplest is to put them into a single dict.

```python
import numpy as np

# Define three objects
# dict
data = {
    "name": "Alice",
    "age": 30,
    "scores": [95, 88, 76]
}
# array
arr = np.array([1.,2.,5.,-7.])
# list
list = ['dog','cat','horse']

# Now bundle them together
bundle = {
    'data':data,
    'array':arr,
    'list':list
}

# Write them to a file
f = open('bundle.pkl','wb')
pickle.dump(bundle,f)
f.close()
```

```python
# Now read them in

f = open('bundle.pkl','rb')

load_bundle = pickle.load(f)

for key in load_bundle:
    print(key,'=',load_bundle[key])
```
