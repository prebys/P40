# Demonstration of SymPy

Make the expression
$$
x^3+2x^2+3x+1
$$

And then do some manipulations with it.

```python
import sympy as sp
from IPython.display import display, Math
%matplotlib inline
import matplotlib.pyplot as plt


# Declare symbolic variables
x = sp.Symbol('x')

# Define a symbolic expression
expr = x**3+2*x**2+3*x+1
# Print expression
print("\nRaw expression:")
print(expr)
# Convert expression to Latex
print("\nLaTeX expression:")
print(sp.latex(expr))
# print formatted in LaTex
print("\nFormatted LaTex expression:")
display(Math(sp.latex(expr)))

print("\nValue (x=2):")
expr_sub = expr.subs({x: 2})
print(expr_sub)

```

# Plotting functions

```python
from sympy import plot
plot(expr, (x, -5, 5), title="My Expression")
plt.show()
```

# Derivatives and Integrals

```python
# Differentiate
dexpr = sp.diff(expr, x)

# Integrate
iexpr = sp.integrate(expr, x)

print("Derivative:")
display(Math(sp.latex(dexpr)))

print("Integral:")
display(Math(sp.latex(iexpr)))
```

# Find roots
Note: roots must be analytically solvable.

All  polynomials have explicit roots, but they're ugly!

```python
# Solve an equation
solution = sp.solve(expr, x)

for s in solution:
    simp = sp.simplify(s)  # Do our best to simplify them
    display(Math(sp.latex(simp)))
```

# Evaluate the roots

```python
# Yikes!  Let's evaluate these
for i,s in enumerate(solution):
    print(f'\nRoot {i}:')
    display(Math(sp.latex(s.evalf())))
    test = expr.subs({x:s.evalf()})
    print('Test:',sp.simplify(test))
```

# Can solve transcendental functions numerically

```python
# Numerical solver

expr = x-sp.cos(x)

display(Math(sp.latex(expr)))

print (sp.nsolve(expr,x,1.))
```

# Two-body momentum conservation

Solve for $p$ in
$$
\sqrt{p^2+m_1^2} + \sqrt{p^2+m_2^2} = M
$$

```python
# Define symbols
M = sp.Symbol('M')
p = sp.Symbol('p')
m1 = sp.Symbol('m_1')
m2 = sp.Symbol('m_2')

# Turn it into a root problem
expr = sp.sqrt(p**2+m1**2) + sp.sqrt(p**2+m2**2) - M
# Check
display(Math(sp.latex(expr)))
```

```python
# Solve
solution = sp.solve(expr,p)
# Note, this will find both the positive and negative root!
for s in solution:
    display(Math(sp.latex(sp.simplify(s))))    # Not quite the same format, but equivalent
```

# Matrix multiplication

```python
from sympy import Matrix

# Define two matrices
A = Matrix([[1, 2],
            [3, 4]])

B = Matrix([[5, 6],
            [7, 8]])

# Multiply the matrices
C = A * B

print("Matrix A:")
display(Math(sp.latex(A)))

print("\nMatrix B:")
display(Math(sp.latex(B)))

print("\nMatrix A * B:")
display(Math(sp.latex(C)))

Ainv = A.inv()

print("\nMatrix Ainv")

display(Math(sp.latex(Ainv)))


T = Ainv*A

print("\nMatrix A * Ainv")

display(Math(sp.latex(T)))
```

# FODO Cell Example

```python
# Symbolic matrices
# Half length of FODO cell and focal length of quadrupole
L,f = sp.symbols('L f')

# Focusing quad
QF = Matrix([[1, 0],
            [-1/(2*f), 1]])
# Defocusing quad
QD = Matrix([[1, 0],
            [+1/(2*f), 1]])

# Drift
D = Matrix([[1, L],
            [0, 1]])

print("\nMatrix QF:")
display(Math(sp.latex(QF)))

print("\nMatrix QD:")
display(Math(sp.latex(QD)))

print("\nMatrix D:")
display(Math(sp.latex(D)))

# Build the FODO cell

M = QF*D*QD*QD*D*QF

print("\nMatrix M:")
display(Math(sp.latex(M)))

M = sp.simplify(M)
print("\nMatrix M (simplified):")
display(Math(sp.latex(M)))

```
