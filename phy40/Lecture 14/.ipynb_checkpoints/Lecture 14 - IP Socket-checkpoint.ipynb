# Asychrononous IP communicationab

This will open a socket and wait for a message from an external process

```python
import asyncio
import socket

async def wait_for_message(port=50007):
    # Create a non-blocking socket
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    s.bind(("0.0.0.0", port))
    s.listen(1)
    s.setblocking(False)

    print(f"Waiting for connection on port {port}...")

    loop = asyncio.get_running_loop()

    # Wait for a client to connect
    conn, addr = await loop.sock_accept(s)
    print("Connected by", addr)

    # Wait for data from client
    data = await loop.sock_recv(conn, 1024)
    print("Received:", data.decode())

    conn.close()
    s.close()

# Start the task
task = asyncio.create_task(wait_for_message())

```

```python
print("Notice how this is not blocking other processes...")
```
