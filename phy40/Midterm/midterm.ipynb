```python
# Scope Example
x = 1
y = 2
# Define a subroutine
def sub():
    y = 3
    print(x)      # Print statement 1
    print(y)      # Print statement 2
# Execute the subroutine
sub()
print(x)          # print statement 3
print(y)          # print statement 4
```
