# Initializing variables and default variable types

```python
# Python will try to guess the variable types
i = 5
j = 5.
x = 2.2
print(type(i),i)
print(type(j),j)
print(type(x),x)
```

```python
# On the other hand, you can explicitly cast variables
i = float(5)
x = int(2.7)  
print(type(i),i)
print(type(x),x)  # Note that it truncated 2.7 rather than rounding it!
```

```python
# Boolean variables
a = True
print(type(a))
b = False
c = a and b
d = a or b
print(c)
print(d)
```

# Formatted printing in Python
For simple things, you don't always need to format

```python
x = 5.5
i = 2
s = "Hello"
print(s,", x = ",x,", i = ",i)
```

On the other hand, Python supports three different ways to format text.

If you're used to C, you can do "old style" formatting, which is like printf() (this is the one I'm most confortable with)

```python
# "old style" formatting.  Based on C/C++/Java "printf()"
x = 5.
i = 10
s = "Hello"
print("The floating point is %.2f, the integer is %d, and the string is %s"%(x,i,s))
```

It also supports "f-string formatting"

```python
print(f"The floating point is {x:.2f}, the integer is {i}, and the string is {s}")
```

Finally, you can use the str.format() method

```python
print("The floating point is {:.2f}, the integer is {}. amd the string is {}".format(x,i,s))
```

# Keyed arrays, AKA "dicts"

Keyed arrays are arrays that are referenced with names rather than indices

```python
# Can do it like this
mydict = {'University':'Davis','Department':'Physics','Start Year':2019}
for key in mydict:
    print("Key = ",key,", value = ",mydict[key],", type = ",type(mydict[key]))
```

```python
# Can also do it on the fly
mydict={} # Create empty keyed array
# fill in the elements
mydict['name']='Eric'
mydict['age']=62
mydict['state']='California'
for key in mydict:
    print("Key = ",key,", value = ",mydict[key],", type = ",type(mydict[key]))
```

# Mutable vs. Immutable

Simple variables are "immutable", meaning they can't be changed

```python
# Example
a = 13   #This actually creates and integer object for "13" and points a to it!
b = a   #This creates a pointer to the same object
print("a = ",a)
print("b = ",b)
print("Location of a = ",id(a))  # id() returns the memory location of an object
print("Location of b = ",id(b))
print("a is b = ",(a is b))   # Are the two objects the *same*
```

```python
# However, if we change a
a = 7  # This actually creates a *new* variable for 7 and points a to it
print("a = ",a)
print("b = ",b)
print("Location of a = ",id(a))
print("Location of b = ",id(b))
print("a is b = ",(a is b))   # Are the two objects the *same*
```

```python
# Now I'll chnage a back to the value of b
c = 7  # this will point to the "7" object, just like "a"
print("a = ",a)
print("c = ",c)
print("Location of a = ",id(a))
print("Location of c = ",id(b))
print("a is c = ",(a is c))   # Are the two objects the *same*
```

# Spinning through lists or tuples

```python
mylist = ['dog','cat',7,5.2,'green',9,10,11,12,'running','out','of','ideas']
# print the whole list
print(mylist)
```

```python
# iterate through the elements
for element in mylist:
    print(element)
```

```python
# Access it via indices
for i in range(len(mylist)):
    print(mylist[i])
```

```python
# Special addressing tools
print(mylist[-1])  # last element
print(mylist[:4])  # first four elements
print(mylist[-4:])  # last 4 elements
print(mylist[2:6]) # elements [2] to [5]
print(mylist[1:8:2]) # elements 1 to 7, incrementing by 2 each time
```

```python
# Note the important difference here
newlist = mylist
print(id(mylist))
print(id(newlist))  # two pointers to the same list!
```

```python
newlist = mylist[:]
print(id(mylist))
print(id(newlist))  # creates a new list! Same as mylist.copy()
```
