```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# nth order taylor sine function
def taylor_trig(x,n):
    sin = 0.
    cos = 0.
    for i in range(n):
        if i%2 == 1:
            sin += (-1)**((i-1)/2)*x**i/math.factorial(i)
        else:
            cos += (-1)**(i/2)*x**i/math.factorial(i)
    return sin,cos
```

```python
print(taylor_trig(np.pi/4.,3))
```

```python
theta = np.linspace(-2*math.pi,2*math.pi,100)
```

```python
x,y = taylor_trig(theta,3)
```

```python
print(x)
```

```python
plt.plot(theta,np.sin(theta))
plt.plot(theta,taylor_trig(theta,3)[0])
plt.plot(theta,taylor_trig(theta,7)[0])
plt.plot(theta,taylor_trig(theta,13)[0])
plt.ylim(-1.5,1.5)
plt.grid()
plt.show()
```

```python
plt.plot(theta,np.cos(theta))
plt.plot(theta,taylor_trig(theta,3)[1])
plt.plot(theta,taylor_trig(theta,7)[1])
plt.plot(theta,taylor_trig(theta,13)[1])
plt.ylim(-1.5,1.5)
plt.grid()
plt.show()
```
