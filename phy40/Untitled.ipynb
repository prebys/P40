```python
def fib(n):
    s = []
    if n > 0:
        s += [0]
    if n > 1:
        s += [1]
    for i in range(2,n):
        s += [s[i-1]+s[i-2]]
    return s
```

```python
print(fib(20))
```

```python
def fib2(n):
    s = None
    if n > 0:
        s = np.array([0])
    if n > 1:
        s = np.append(s,1)
    for i in range(2,n):
        s = np.append(s,s[i-1]+s[i-2])
    return s
```

```python
print(fib2(20))
```
