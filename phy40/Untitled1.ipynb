```python
import sympy as sp
```

```python
x = sp.Symbol('x')
```

```python
expr = sp.exp(sp.sin(2*x))
```

```python
print(expr)
```

```python
diff = sp.diff(expr)
print(diff)
```

```python
from IPython.display import Markdown
ddx = r"\frac d{dx}"
Markdown(f"$${ddx}{sp.latex(expr)}={sp.latex(diff)}$$")
```
