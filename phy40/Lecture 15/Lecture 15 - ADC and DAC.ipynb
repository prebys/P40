# Examples of ADC and DAC code

```python
# DAC routine
def DAC(D,Vlo,Vhi,nbit):
    if D<0 or D>(2**(nbit)-1):
        print("D out of range!")
        return 0.
    else:
        A = Vlo + (Vhi-Vlo)*D/2**nbit
        return A
```

```python
# ADC routine
def ADC(A,Vlo,Vhi,nbit):
    if A<Vlo:
        return 0
    elif A>=Vhi:
        return 2**nbit-1
    else:
        D = int((A-Vlo)/(Vhi-Vlo)*2**nbit)
        return D
```

```python
print(DAC(125,0.,5.,8))
```

```python
print(ADC(5.,-10.,10.,16))
```
