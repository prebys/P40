```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def parkmiller(i):
    a = 7**5
    c = 0
    M = 2**31-1
    return (a*i+c)%M
```

```python
i = 1
for n in range(10000):
    i = parkmiller(i)
print(i)
```

```python
i = 93939399
def callpark():
    global i
    i = parkmiller(i)
    return(i)
```

```python
s = np.array([callpark() for n in range(10000)])
```

```python
s = s/(2**31-1)
```

```python
plt.hist(s,bins=20)
plt.show()
```
