This will fill the tables used in the object oriented example

```python
import pandas as pd
import numpy as np
```

```python
def main():
    # Create the table of PMTs
    pmts = create_PMT_table()
    pmts.to_csv("detector.csv")  # Write it to a file
    # Generate two sets of hits
    # First set
    hits = create_hits(20)
    hits.to_csv("hits1.csv")      # Write to a file
    # Second set
    hits = create_hits(20)
    hits.to_csv("hits2.csv")      # Write to a file
    
```

```python
# Create a table defining the PMTs.  It will create a 10x10 array of PMTs, separated by 10cm, 10m downstream
# The gains offsets will be distributed around 10 and the gains will be distributed about .01.
def create_PMT_table():
    ntubes = 100
    pmts = pd.DataFrame(columns=["ID", "x", "y", "z", "offset","gain"])
    for i in range(ntubes):
        ID = i+1
        x = i%10*.1+.05
        y = i//10*.1+.05
        z = 10.
        offset = np.random.normal(10.,10.)  # distributed about 0 with a sigma of 10.
        gain = np.random.normal(.01,.002) # Distributed about .05 with a sigma of .001
        pmts.loc[i] = [ID,x,y,z,offset,gain]
    return pmts
        
```

```python
# generate a list of tube IDs and hits
def create_hits(nhits):
    hits = pd.DataFrame(columns=["ID", "raw"])
    for i in range(nhits):
        ID = np.random.randint(1,101)
        raw = np.random.normal(9000.,1000.)
        hits.loc[i] = [ID,raw]
    return hits
```

```python
# Now execute the main loop
main()
```
