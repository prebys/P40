# Fibonacci Series

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Two ways of generating whole sequences

```python
# Generate an n-element Fibonacci Sequence as a list
def fiblist(n):
    seq = [0]             # begin the list with the first number
    if n==1:
        return seq        # Return if we only asked for one value
    seq += [1]            # Add the second
    for i in range(2,n):  # If n<=2, this loop will not execute
        seq += [seq[i-1] + seq[i-2]]  # Increase list by appending the sum of the previous previous terms
    return seq
```

```python
# Generate an n-element Fibonacci Sequence as an array
import numpy as np
def fibarray(n):
    seq = np.empty(n,dtype=np.uint64)     # Make it unsigned integers
    seq[0] = 0                      # begin the list with the first number
    if n==1:
        return seq                  # Return if we only asked for one value
    seq[1] = 1                      # Add the second
    for i in range(2,n):            # If n<=2, this loop will not execute
        seq[i] = seq[i-2]+seq[i-1]  # Increase list by appending the sum of the previous previous terms
    return seq
```

```python
# Print the first 10 elements
print("fiblist(10)  = ",fiblist(10))
print("fibarray(10) = ",fibarray(10))
```

```python
# Print the last elements for a large sequence
n=100
print(f"Last element of fiblist({n}) = {fiblist(n)[-1]}")
print(f"Last element of fibarray({n}) = {fibarray(n)[-1]}")
```

Fibonacci Sequence and the Golden ratio

```python
# Generating the golden ratio
arr = fibarray(15)
ratio = np.array([(arr[i]/arr[i-1]) for i in range(2,15)])  # Powerful way to fill an array
```

```python
x = np.arange(2,15)
gold = np.full(len(x),1.618033988749)

plt.plot(x,ratio,"b.-",ms=15,label=r"$F_n/F_{n-1}$")
plt.plot(x,gold,"r--",label="Golden Ratio (1.6180)")
plt.xlabel("n")
plt.ylabel("Value")
plt.legend()
plt.show()
```
