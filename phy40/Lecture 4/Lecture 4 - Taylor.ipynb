# Taylor Expansions

exponential, sine, and cosine

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Expand an exponent to the nth order.  
def expseries(x,n):
    val = np.ones(len(x))       # Create it as an array to handle array arguments    
    for i in range(1,n+1):      # Note! Need to get order right
        val += x**i/math.factorial(i)
    return val
```

```python
x = np.linspace(0.,2.,100)
plt.xlim(0.,2.)
plt.plot(x,np.exp(x),label = r"$e^x$")
y = expseries(x,1)
plt.plot(x,y, "--", label = r"1$^{st}$ Order Expansion")
y = expseries(x,2)
plt.plot(x,y, "--",label = r"2$^{nd}$ Order Expansion")
y = expseries(x,3)
plt.plot(x,y, "--",label = r"3$^{rd}$ Order Expansion")
plt.legend()
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
x = np.linspace(0.,2.,100)
plt.xlim(0.,2.)
plt.ylim(0.,1.)
plt.plot(x,np.exp(-x),label = r"$e^{-x}$")
y = expseries(-x,1)
plt.plot(x,y, "--", label = r"1$^{st}$ Order Expansion")
y = expseries(-x,2)
plt.plot(x,y, "--",label = r"2$^{nd}$ Order Expansion")
y = expseries(-x,3)
plt.plot(x,y, "--",label = r"3$^{rd}$ Order Expansion")
plt.xlabel('x')
plt.legend()
plt.grid()
plt.show()
```

```python
# nth order taylor sine function
def sin_series(x,n):
    sin = np.zeros(len(x))
    for i in range(n+1):
        if i%2 == 1:
            sin += (-1)**((i-1)/2)*x**i/math.factorial(i)
    return sin
```

```python
theta = np.linspace(-2*math.pi,2*math.pi,100)

plt.plot(theta,np.sin(theta),label=r"sin($\theta$)")
plt.plot(theta,sin_series(theta,1),"--", label = r"1$^{st}$ Order Expansion")
plt.plot(theta,sin_series(theta,7),"--", label = r"7$^{th}$ Order Expansion")
plt.plot(theta,sin_series(theta,13),"--", label = r"13$^{th}$ Order Expansion")
plt.ylim(-1.5,1.5)
plt.legend()
plt.xlabel(r'$\theta$')
plt.grid()
plt.show()
```

```python
# nth order taylor sine function
def cos_series(x,n):
    sin = np.zeros(len(x))
    for i in range(n+1):
        if i%2 == 0:
            sin += (-1)**((i)/2)*x**i/math.factorial(i)
    return sin
```

```python
theta = np.linspace(-2*math.pi,2*math.pi,100)

plt.plot(theta,np.cos(theta),label=r"cos($\theta$)")
plt.plot(theta,cos_series(theta,2),"--", label = r"2$^{nd}$ Order Expansion")
plt.plot(theta,cos_series(theta,8),"--", label = r"8$^{th}$ Order Expansion")
plt.plot(theta,cos_series(theta,14),"--", label = r"14$^{th}$ Order Expansion")
plt.ylim(-1.5,1.5)
plt.legend()
plt.xlabel(r'$\theta$')
plt.grid()
plt.show()
```
